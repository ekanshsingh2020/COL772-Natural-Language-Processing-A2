```python
import numpy as np
import json
import pandas
import transformers
from gensim.models import Word2Vec 
from gensim.models import FastText
from sklearn.metrics.pairwise import cosine_similarity
import gensim.downloader
import nltk
```

```python
questions = []
tables = []
actual_col = []
label_cols = []

# Read JSON data from file line by line
with open('data/A2_train.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        # Parse JSON data from each line
        parsed_data = json.loads(line)
        
        # Extract required fields and append to respective lists
        # append the parsed_data[question] but all in lower case
        
        questions.append(parsed_data['question'])
        
        tables.append(parsed_data['table'])
        label_cols.append(parsed_data['label_col'])
        # Extract actual column names from table
        actual_col.append(list(parsed_data['table']['cols']))

print('Number of questions:', len(questions))
print('Number of tables:', len(tables))
print('Number of label columns:', len(label_cols))
print('Number of actual columns:', len(actual_col))
```

```python
## convert the list of questions into a two dimensional array of words
# question_array = []
# for question in questions:
#     question_array.append(question.split())
# print (question_array)
token_array = []
for question in questions:
    # print(question)
    tokens = nltk.word_tokenize(question)
    # make all tokens lower case
    tokens = [token.lower() for token in tokens]
    token_array.append(tokens)
print(token_array[0])
```

```python
print (list(gensim.downloader.info()['models'].keys()))
```

```python
temp_list = []
temp_list.append('glove-twitter-25')
```

```python
for model in temp_list:
    word2vec_model = gensim.downloader.load(model)
        # for every question, convert each word to a vector and then sum the vectors to get a single vector for the question
    question_vectors = []
    for question in token_array:
        question_vector = np.zeros(25)
        for word in question:
            try:
                question_vector = np.add(question_vector, word2vec_model[word])
            except:
                # print (word)
                pass
        question_vectors.append(question_vector)
    
    similar_columns = []
    for i in range(len(question_vectors)):
        # calculate the cosine similarity between the question vector and each column vector
        # make a np array 
        similarity =[]
        for col in actual_col[i]:
            sim = 0
            temp_col = col.lower()
            temp_col = temp_col.replace(" ", "")
            try :
                sim = cosine_similarity([question_vectors[i]], [word2vec_model[temp_col]])
            except:
                sim = -1
            similarity.append(sim)
        maxval = max(similarity)
        idx=-1;
        for j in range (len(similarity)):
            if similarity[j]==maxval:
                idx = j
        similar_columns.append(actual_col[i][idx])
        # print (similarity)
        # print("=====================================")
    # print(similar_columns)
        
    ans =0
    for i in range(len(similar_columns)):
        print (similar_columns[i])
        print (label_cols[i][0])
        if similar_columns[i] == label_cols[i][0]:
            ans+=1

    accuracy = ans/len(similar_columns)
    print (model)
    print (ans)
    print(accuracy)
    with open('results.txt', 'a') as file:
        file.write(model)
        file.write("\n")
        file.write(str(ans))
        file.write("\n")
        file.write(str(accuracy))
        file.write("\n")
        file.write("=====================================")
        file.write("\n")
```
